# Forward Plume (Influence Domain)

Traditional Navier-Stokes solver given the inital guess of source location could provide a influence domain, which is the forward expanding plume.

```python
t_ind = 5000
k_ind = 1
var_ind = 'theta'
timestep = np.linspace(0, t_ind, 201)
result_dir = '/home/zyou6474/tasks/channel_flow'
dims = [128, 128, 64]
domain = [2*np.pi, np.pi, 1]

ldata = lesgo_data(domain, dims, result_dir, ntheta=3)
```

```python
ldata.read_data(t_ind)
```

```python
fig, ax = plt_utils.contour_channel(ldata.coords, ldata.data[var_ind][k_ind], vmin=0, vmax=1)
```

```python
plt_utils.generate_gif(timestep, ldata, fig_dir, var_ind, k_ind, job_mark+date_mark, vmin=0, vmax=1)
```
