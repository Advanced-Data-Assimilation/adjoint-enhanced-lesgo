```python
import pyutils.plot_utils as plt_utils
from turb.lesgo_utils import lesgo_data
from matplotlib import cm, colors
import matplotlib.pyplot as plt
import numpy as np

fig_dir = r'/home/zyou6474/Projects/ERI/notebooks/figures'

job_mark = 'adjoint_steady_'

from datetime import datetime
date_mark = datetime.today().strftime('%Y-%m-%d')
```

# Adjoint Field of Sensors

Set IC to be constant zero and source at different height at each x grid.

```python
root_dir = '/home/zyou6474/tasks/adjoint_channel_flow'
dims = [128, 128, 64]
domain = [2*np.pi, np.pi, 1]
tspan = 5000

# Sensor Location
sx = (5/3*np.pi, 5/3*np.pi, 5/3*np.pi)
sy = (1/2*np.pi, 1/2*np.pi, 1/2*np.pi)
sz = (1/2, 1/3, 2/3)

sensor_locs = (sx, sy, sz)


ldata = lesgo_data(domain, dims, root_dir, ntheta=3)
ldata.set_adjoint(True, adjoint_dir=root_dir, fmt_ntheta='%.3i')


# Write np.roll to write IC files
ldata.sensor_init_(sensor_locs)

ldata.sensor_field()
```

```python
ldata.read_inputs()
print('theta sum is ', ldata.data['theta.IC'].sum())
print('theta max is ', ldata.data['theta.IC'].max())
```
